```python
import pandas as pd
```

```python
la_df = pd.read_csv("./la_listings.csv")
ny_df = pd.read_csv("./ny_listings.csv")
la_df["state"] = "LA"
ny_df["state"] = "NY"

df_list = [la_df, ny_df]
```

```python
all_df = pd.concat(df_list, axis=0)
all_df
```

```python
all_df.columns
```

```python
f = open("./untitled.txt")
lines = f.readlines()

temp_list = []
for line in lines:
    if line == lines[-1]:
        temp_list.append(line)
    else:
        temp_list.append(line[:-1])
    
temp_list
```

```python
new_list = [i for i in temp_list if i in list(all_df.columns)]
```

```python
new_list
```

```python
all_df = all_df.loc[:, new_list]
all_df
```

```python
all_df = all_df.dropna()
all_df
```

```python
all_df["host_response_rate"] = all_df["host_response_rate"].apply(lambda x: int(x[:-1]))
all_df["price"] = all_df["price"].apply(lambda x: (x[1:-3]).replace(",", ""))
all_df
```

```python
all_df.to_csv("./preprocessing_dataframe.csv")
```
